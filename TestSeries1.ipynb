```python
import numpy as np
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn import linear_model 
def lin_tests(X_train, X_test, y_train, y_test):

    print("Linear Tests Started")
 
    lin_model = linear_model.LinearRegression().fit(X_train,y_train)
    results = evaluate_model(lin_model, X_test, y_test)
    dl.save_info(str(lin_model),results)

    lin_model = linear_model.Ridge(alpha=.5)
    lin_model.fit(X_train,y_train)
    results = evaluate_model(lin_model, X_test, y_test)
    dl.save_info(str(lin_model),results)


    lin_model = linear_model.Lasso(alpha=0.1)
    lin_model.fit(X_train,y_train)
    results = evaluate_model(lin_model, X_test, y_test)
    dl.save_info(str(lin_model),results)

    lin_model = linear_model.MultiTaskLasso(alpha=0.1)
    lin_model.fit(X_train,y_train)
    results = evaluate_model(lin_model, X_test, y_test)
    dl.save_info(str(lin_model),results)

    lin_model = linear_model.MultiTaskElasticNetCV(cv=5, random_state=0, max_iter=1000000)
    lin_model.fit(X_train,y_train)
    results = evaluate_model(lin_model, X_test, y_test)
    dl.save_info(str(lin_model),results)

    lin_model = linear_model.Lars(n_nonzero_coefs=1)
    lin_model.fit(X_train,y_train)
    results = evaluate_model(lin_model, X_test, y_test)
    dl.save_info(str(lin_model),results)

    lin_model = linear_model.LassoLars(alpha=.1)
    lin_model.fit(X_train,y_train)
    results = evaluate_model(lin_model, X_test, y_test)
    dl.save_info(str(lin_model),results)
    
    print("Linear Tests Complete")
    print()
```
